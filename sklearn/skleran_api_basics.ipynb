# Guiding Principles

1. Consistency: Common interface with a limited set of methods and consistent documentation.

2. Inspection: All parameter values are exposed as public attributes.

3. Limited Object Hierarchy: Only algorithms are represented as Python classes; datasets use standard formats (NumPy arrays, Pandas DataFrames, SciPy sparse matrices).

4. Composition: Tasks are expressed as sequences of fundamental algorithms.

5. Sensible Defaults: Default values are provided for user-specified parameters.

# Basic steps of the API
1. Choose Model Class: Import the appropriate estimator class from Scikit-Learn.
2. Set Model Hyperparameters: Instantiate the class with desired values.
3. Arrange Data: Data should be organized into a features matrix and target vector.
4. Fit the Model: Use the fit() method of the model instance to train the model on your data.
5. Apply the Model: 
    (a) Supervised Learning: Use the predict() method to predict labels for unknown data.
    (b) Unsupervised Learning: Use transform() or predict() methods to infer properties of the data.

# Example: Fit a Linear Regression model


```python
# Create a sample data
import numpy as np
rng = np.random.RandomState(42)

x = 10*rng.rand(50)
x
```

```python
y = 2 * x -1+rng.randn(50)
y
```

```python
# plot sample data
import matplotlib.pyplot as plt

plt.scatter(x,y)
```

```python
# 1. Choose model class: each class has each ML model
# 2. Choose the hyperparameters
from sklearn.linear_model import LinearRegression

# set hyperparameters with instance initiation
model = LinearRegression(fit_intercept=True)
```

```python
model
```

```python
# 3. Arrage data X: Fetaure matrix and y: Target array
x.shape  # this is an array
```

```python
X = x[:, np.newaxis]   # change to matrix
```

```python
X.shape
```

```python
y   # we need array only as target array
    # No need to change
```

```python
# 4. Fit the model
model.fit(X,y)   
```

- This stored all the data in its rspective attributes (which are model specific).

```python
# Access rsult using attributes
model.coef_   # SLope of fitted line
```

```python
model.rank_
```

```python
model.intercept_   # line intercept
```

```python
model.n_features_in_
```

```python

# 5. Predict labels for unknown data.

# Create test data for now

test_data = np.linspace(-1,11)
test_data
```

```python
Xfit = test_data[:, np.newaxis]
Xfit
```

```python
yfit = model.predict(Xfit)
```

```python
yfit
```

```python
plt.plot(Xfit,yfit)
plt.scatter(X, y)
plt.show
```
